## Random Forest & Extra Tree Model

```python
X = new_df.drop(columns=['size_code', 'size'], axis=1)
y = new_df['size_code']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42, stratify=y)
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
et = ExtraTreesClassifier(n_estimators=100)
```

```python
cross_val_score(rf, X_train, y_train, cv=5).mean()
```

```python
cross_val_score(et, X_train, y_train, cv=5).mean()
```

### GridSearch of Random Forest

```python
rf_params = {
    'n_estimators': [100, 150, 200],
    'max_depth': [None, 1, 2, 3, 4, 5],
}
gs = GridSearchCV(rf, param_grid=rf_params, cv=5)
gs.fit(X_train, y_train)
print(gs.best_score_)
gs.best_params_
```

```python
gs.score(X_train, y_train)
#overfit
```

```python
gs.score(X_val, y_val)
```

```python
rf_predictions = gs.predict(X_test)
```

## Boosting Models

```python
X = new_df.drop(columns=['size_code', 'size'], axis=1)
y = new_df['size_code']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
AdaBoostClassifier(base_estimator=DecisionTreeClassifier())
```

```python
GradientBoostingClassifier()
```

```python
knn_pipe = Pipeline([
    ('ss', StandardScaler()),
    ('knn', KNeighborsClassifier())
])
```

```python
vote = VotingClassifier([
    ('ada', AdaBoostClassifier(base_estimator=DecisionTreeClassifier())),
    ('gb', GradientBoostingClassifier()),
    ('tree', DecisionTreeClassifier()),
    ('knn_pipe', knn_pipe)
])

params = {
    'gb__n_estimators': [100],
    'ada__base_estimator__max_depth': [2, 3, 4],
    'knn_pipe__knn__n_neighbors': [4, 5]
}
gs = GridSearchCV(vote, param_grid=params)
gs.fit(X_train, y_train)
print(gs.best_score_)
gs.best_params_
```
